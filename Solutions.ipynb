```python
import pandas as pd 
import matplotlib.pyplot as plt

df = pd.read_csv('C:\\Users\\KIIT\\Documents\\Python Scripts\\Data Manipulation case study(module 6)\\BigMartSalesData.csv')

df_grouped_year = df.query("Year == 2011").filter(["Month", "Amount"]).groupby(["Month"], as_index=False).sum()

x = df_grouped_year["Month"]
y = df_grouped_year["Amount"]

plt.plot(x, y)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Month vs sales")

plt.grid()
plt.show()
```

Here we can see the sale incresed almost 11th month but after that it was started falling down

And in 2nd month the lowest sale occurs.

```python
x = df_grouped_year["Month"]
y = df_grouped_year["Amount"]

plt.bar(x, y)
plt.xlabel("Month")
plt.ylabel("Total Sales")
plt.title("Month vs sales")
plt.grid()
plt.show()
```

Since the line charts are used to track the changes over the short and long periods of time and whenever a small change occurs line graph is much better to use than bar graphs because line graphs used to compare the changes over the same period of time for more than one group.However, when we are trying to measure the changes over time bar graphs are best option.

```python
df_grouped_country = df.query("Year == 2011").filter(
    ["Country", "Amount"]).groupby(["Country"], as_index=False).sum()

plt.pie(df_grouped_country["Amount"],
        labels=df_grouped_country["Country"],
        autopct='%1.1f%%',
        shadow=True,radius=3)

plt.tight_layout()
plt.show()
```

Here clearly we can see that United Kingdom has highest towards the sales.
